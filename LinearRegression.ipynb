#### **Theory:**
Linear regression analysis is used to predict the value of a variable based on the value of another variable. The variable you want to predict is called the dependent variable. The variable you are using to predict the other variable's value is called the independent variable.<br>
Linear-regression models are relatively simple and provide an easy-to-interpret mathematical formula that can generate predictions. Linear regression can be applied to various areas in business and academic study.<br>



Import the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('homeprice.csv')
df
```

```python
plt.scatter(df.Area, df.Price)
```

```python
x = df.iloc[:, 0].values.reshape(-1,1)
x
y = df.iloc[:, 1].values.reshape(-1, 1)
y
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

Regression Line

```python
y_pred = reg.predict(x)
```

```python
plt.plot(x, y_pred, color='black')
plt.scatter(df.Area, df.Price, marker='+')
plt.show()
```

**Formula:** Y = β0 + β1X + e
<br>
Y = Dependent variable / Target variable<br>
β0 = Intercept of the regression line <br>
β1 = Slope of the regression line which tells whether the line is increasing or decreasing<br>
X = Independent variable / Predictor variable<br>
e = Error
